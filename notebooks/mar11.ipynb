```python
import os
import altair as alt
import pandas as pd
```

```python
df = pd.read_csv('../data/ALL_CITIES.csv')
```

```python
df_sea = df[df['city'] == 'Seattle']

df_sea.shape
```

```python
sea_scatter = alt.Chart(df).mark_circle(opacity=0.1).encode(
  x=alt.X('difference_min_temp'),
  y='difference_max_temp',
  tooltip=['city', 'date'],
  facet=alt.Facet('city:N', columns=3)
).properties(
  width=120,
  height=120
)

sea_scatter
```

```python
sea_line = alt.Chart(df).mark_circle().encode(
  x='date:T',
  y='difference_max_temp:Q',
  tooltip=['city', 'date:T', 'actual_max_temp', 'actual_min_temp']
).properties(
  width=800,
  height=300
)

sea_line
```
